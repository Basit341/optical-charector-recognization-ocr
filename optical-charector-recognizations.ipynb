```python
image = Image.open('/kaggle/input/questions/download.png')
image = np.array(image)
image
```

```python
import easyocr
reader=easyocr.Reader(['en'],gpu=True)
results=reader.readtext(image)
print(results[0][1])
print(results[1][1])
print(results[2][1])
print(results[3][1])
print(results[4][1])
print(results[5][1])
print(results[6][1])
print(results[7][1])
```

```python
easy=pd.DataFrame(results,columns=['bbox','text','conf'])
```

```python
import keras_ocr
pipeline=keras_ocr.pipeline.Pipeline()
results=pipeline.recognize(image)
```

```python
from tqdm import tqdm
reader=easyocr.Reader(['en'],gpu=True)
dfs=[]
for img in tqdm(image):
    results=reader.readtext(image)
    image_df=pd.DataFrame(results,columns=['bbox','text','coef'])
    dfs.append(image_df)
```

```python
easy_ocr=pd.concat(dfs)
easy_ocr
```
